#**Problema de Negocio**

Algoritmo capaz de analizar caracteristicas de propiedad para sugerir tarifas diarias para mejores ganancias en alta demanda.

##1. Entendiendo el problema

```python
import pandas as pd
```

```python
datos = pd.read_json('/content/datos_hosting.json')
datos.head()
```

```python
#Sobreescribiendo la variable datos, con una normalizacion
#Hace una estructuracion de los datos en un dataframe
datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()
```

##2. Datos núméricos

```python
#Explorando los datos
columnas = list(datos.columns)
columnas
```

```python
#Genera un nuevo registro de cada columna que tiene listas,
# tendrán el mismo indice
datos = datos.explode(columnas[3:])
datos
```

```python
#Resetea el indice en el mismo dataframe y le asigna nuevos
datos.reset_index(inplace=True, drop=True)
datos.head()
```

```python
#Coloca object porque no sabe que tipo de dato es
datos.info()
```

```python
#Colocando el tipo de dato correcto en la columna
import numpy as np
```

```python
#Le cambia el tipo de dato a la columna como entero con numpy
#Si en algún valor no hay enteros, no lo cambiará
datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
```

```python
datos.info()
```

```python
columnas_numericas = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
#Le cambia el tipo a todas las columnas de la lista
datos[columnas_numericas] = datos[columnas_numericas].astype(np.int64)
```

```python
datos.info()
```

```python
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
datos.info()
```

```python
#Transformando (retirando) el string del signo de moneda
#Apply aplica solo con una columna, no con más de una
datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',', '').strip())
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
datos[['cuota_deposito', 'cuota_limpieza']]
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',', '').strip())
```

*Usamos el applymap para cambiar valores en un dataframe completo, ese método mapea todo el dataframe y aplica la función que se le esta dando*

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

Extrayendo valores de texto que pueden influir en el precio (Tokenización)

**Tokenización Simple**

```python
#transforma a string y pone en minusculas
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
datos.head()
```

**Regex**: Busqueda personalizada en un texto, para seleccionar algo en el texto (caracteres especiales, números, etc.)

```python
datos['descripcion_local'][3169]
```

```python
#^ le dice que sustituya todo, excepto lo que se le dice
#en este caso, letras de a a z, números de 0 a 9, guiones y comilla simples
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```

```python
#reemplazar guiones que no tengan una palabra o caracter antes o después
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex=True)
```

```python
datos['descripcion_local'].head()
```

```python
#Tokenización, generando listas de palabras de cada descripción
#El espacio es el caracter que separa las palabras
datos['descripcion_local'] = datos['descripcion_local'].str.split()
```

```python
datos.head()
```

```python
#LA barra es para indicar, y también busca tal...
datos['comodidades'] = datos['comodidades'].str.replace('\{|}|\"', '', regex=True)
```

```python
datos.head()
```

```python
datos['comodidades'] = datos['comodidades'].str.split(',')
```

```python
datos.head()
```

```python
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.lower()
```

```python
datos['descripcion_vecindad'][1]
```

```python
datos['descripcion_vecindad'] = datos['descripcion_vecindad'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
```
